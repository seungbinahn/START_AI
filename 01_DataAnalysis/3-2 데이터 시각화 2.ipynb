```python
import seaborn as sns
import matplotlib.pylab as plt

tips = sns.load_dataset('tips')
tips.head()
```

```python
ax = plt.subplots()
ax = sns.distplot(tips['total_bill'])
ax.set_title('Total Bill Plot')
```

```python
ax = plt.subplots()
ax = sns.distplot(tips['total_bill'], kde = False) # 확률 분포 제거
ax.set_title('Total Bill Plot')
```

```python
ax = plt.subplots()
ax = sns.distplot(tips['total_bill'], hist = False) # 확률 분포 제거
ax.set_title('Total Bill Plot')
```

```python
ax = plt.subplots()
ax = sns.distplot(tips['total_bill'], rug = True) # 확률 분포 제거
ax.set_title('Total Bill Plot')
```

```python
# 산점도 그래프
# 회귀선 표시
ax = plt.subplots()
ax = sns.regplot(x='total_bill', y='tip', data=tips)
ax.set_title('Scatterplot of Total Bill and Tip')
ax.set_xlabel('Total Bill')
ax.set_ylabel('Tip')
```

```python
# 산점도 그래프
ax = plt.subplots()
ax = sns.regplot(x='total_bill', y='tip', data=tips, fit_reg=False) # 회귀선 지우기
ax.set_title('Scatterplot of Total Bill and Tip')
ax.set_xlabel('Total Bill')
ax.set_ylabel('Tip')
```

```python
joint = sns.jointplot(x='total_bill', y='tip', data=tips)
joint.set_axis_labels(xlabel='Total Bill', ylabel='Tip')
joint.fig.suptitle('Joint Plot of Total Bill and Tip', fontsize=10, y=1.03)
```

```python
joint = sns.jointplot(x='total_bill', y='tip', data=tips, kind='hex')
joint.set_axis_labels(xlabel='Total Bill', ylabel='Tip')
joint.fig.suptitle('Joint Plot of Total Bill and Tip', fontsize=10, y=1.03)
```

```python
kde, ax = plt.subplots()
ax = sns.kdeplot(data=tips['total_bill'], data2=tips['tip'],shade=True)
print(type(kde))
print(type(ax))
```

## 바 그래프
* 막대의 높이로 평균(기본값), 검은선으로 신뢰구간을 출력
* estimator 인자로 평균이 아닌 합 출력가능
* x축 데이터는 범주형, y축 데이터는 연속형 데이터 사용
* 순위형 자료에 대한 Counting
* 여러 범주형 변수에 대한 Overlay 쉽게 확인

```python
# 바 그래프
ax = plt.subplots()
ax = sns.barplot(x='time', y='total_bill', data=tips)
# ax = sns.barplot(x='time', y='total_bill', data=tips, estimator = sum)
ax.set_xlabel('Time of day')
ax.set_ylabel('Average total bill')
```

## Box plot
* Five-number Summary : 5가지 주요 통계량이 시각적으로 표현
* 여러 그룹 간 데이터를 비교할 때 유용함
* 데이터의 신뢰구간과 이상치를 빠르게 확인할 수 있음

```python
# box 그래프
ax = plt.subplots()
ax = sns.boxplot(x='time', y='total_bill', data=tips)
ax.set_title('Boxplot of total bill by time of day')
ax.set_xlabel('Time of day')
ax.set_ylabel('Average total bill')
```

```python
# box 그래프
ax = plt.subplots()
ax = sns.violinplot(x='time', y='total_bill', data=tips)
ax.set_title('violinplot of total bill by time of day')
ax.set_xlabel('Time of day')
ax.set_ylabel('Average total bill')
# 빈도수도 알수 있음
```

```python
pair_grid = sns.PairGrid(tips)
pair_grid = pair_grid.map(upper(sns.regplot))
pair_grid = pair_grid.map(lower(sns.kdeplot))
pair_grid = pair_grid.map_diag(sns.regplot))
```

```python
# box 그래프
ax = plt.subplots()
ax = sns.violinplot(x='time', y='total_bill', data=tips, hue='sex')
ax.set_title('violinplot of total bill by time of day')
ax.set_xlabel('Time of day')
ax.set_ylabel('Average total bill')
# 빈도수도 알수 있음
```

```python
facet = sns.FacetGrid(tips, col='time')
facet.map(sns.distplot, 'total_bill',rug=True)
```

```python
# 세로축은 시간으로 나누고
# 가로축은 smoker로 나누고
facet = sns.FacetGrid(tips, col='time',row='smoker',hue='sex')
facet.map(plt.scatter, 'total_bill','tip')
```
