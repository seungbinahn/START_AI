```python
# 2-13
m = np.random.randn(5,5)
print(m)
print('='*10)
print('전체 합 : %.4f' % m.sum())
print('전체 평균 : %.4f' % m.mean())
print('='*10)
c_s = m.sum(axis=0)
c_m = m.mean(axis=0)
for i in range(5):
    print('%d열의 합 : %.4f / 평균 : %.4f' % (i,c_s[i],c_m[i]))

print('='*10)
r_s = m.sum(axis=1)
r_m = m.mean(axis=1)
for i in range(5):
    print('%d행의 합 : %.4f / 평균 : %.4f' % (i,r_s[i],r_m[i]))
```

```python
# 3-1
import numpy as np
import pandas as pd
data = pd.read_csv('data/@preprocessing_data_member01.csv',encoding='utf-8')
print(data.head(n=7))
print(data.tail(n=7))
```

```python
#3-3
#data.info()
data.describe()
```

```python
#3-4
data['성별(replace)'] = data['성별'].replace(['남자','여자'],[1,0])
data[['성별','성별(replace)']]
```

```python
female_data = data[data['성별'] == '여자']
female_data.head()
```

```python
# 3-6
print(pd.pivot_table(data=data, index=['가입연도'], values='총판매액'))
rst = data[data['최종주문연도']==2019].sort_values('총판매액').head(n=5)
rst[['ID','최종주문연도','총판매액']]
```

```python
# 3-7
#table = pd.pivot_table(data=data, index=['최종주문요일','유입경로'], values='총구매횟수', aggfunc='sum')

tb = pd.pivot_table(data=data, index=['최종주문요일'], values='총판매액', aggfunc='sum')
day = tb.sort_values(by=['총판매액','최종주문요일',], ascending=False).index[0]
newdata = data[data['최종주문요일'].isin([day])]
pd.pivot_table(data=newdata, index=['유입경로'], values='총판매액', aggfunc='sum').sort_values(by='총판매액').tail(n=1)

#table.sort_values(by='총구매횟수', ascending=False)
```
