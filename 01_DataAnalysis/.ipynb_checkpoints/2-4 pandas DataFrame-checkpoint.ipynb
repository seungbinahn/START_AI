```python
# 열 추가(serise 추가)
scientists['sex']=['male','male','male','male','female','female','female','female']
scientists.head()
```

```python
# replace 칼럼값 벼경
scientists['sex(replace)'] = scientists['sex'].replace('male',1).replace('female',0)
scientists.head()
```

```python
# replace 칼럼값 벼경
scientists['sex(replace)'] = scientists['sex'].replace({'male':1,'female':0})
scientists.head()
```

```python
# 시간 자료형 처리
born_datetime = pd.to_datetime(scientists['Born'],format='%Y-%m-%d')
print(born_datetime)
```

```python
died_datetime = pd.to_datetime(scientists['Died'],format='%Y-%m-%d')
print(died_datetime)
```

```python
scientists['born_dt'],scientists['died_dt'] = (born_datetime,died_datetime)
scientists
```

```python
# 시계열 자료 계산
scientists['age_days_dt_dt'] = scientists['died_dt'] - scientists['born_dt'] 
scientists.head()
```

```python
# 열 지우기
scientists_dropped = scientists.drop(['Born','Died'],axis=1)
scientists_dropped 
```

```python
# 열 지우기
scientists_dropped2 = scientists.drop([2,7],axis=0)
scientists_dropped2 
```
