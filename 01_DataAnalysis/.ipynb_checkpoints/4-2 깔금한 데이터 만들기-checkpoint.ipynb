```python
# 누락값 0으로 채우기
ebola.fillna(0).iloc[:10,:5]
```

```python
# 누락값 이후의 첫번째 값으로 채우기
ebola.fillna(method='bfill').iloc[:10,:5]
```

```python
# 전후값을 이용하여 중간값으로 채우기
ebola.interpolate().iloc[:10,:5]
```

```python
# 누락값 삭제하기
print(ebola.shape)
ebola_dropna = ebola.dropna()
print(ebola_dropna.shape)
print(ebola_dropna)
```
