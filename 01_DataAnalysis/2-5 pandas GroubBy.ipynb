```python
import numpy as np
import pandas as pd

data = pd.read_csv('data/@preprocessing_data_member01.csv',encoding='utf-8')
data.head()
```

```python
data.groupby('최종주문요일').mean()
```

```python
# mean : 평균
data.groupby('최종주문요일')['총구매횟수'].mean()
```

```python
# sum : 합
data.groupby('최종주문요일')['총구매횟수'].sum()
```

```python
data.groupby(['최종주문요일','성별'])['총구매횟수'].sum()
```

```python
pd.pivot_table(data=data, index='최종주문요일', values=['총구매횟수'])
```

```python
# T 함수 : 
pd.pivot_table(data=data, index='최종주문요일', values='총구매횟수').T
```

```python
pd.pivot_table(data=data, index='최종주문요일', values='총구매횟수', columns='성별').T
```

```python
pd.pivot_table(data=data, index='최종주문요일', values='총구매횟수', aggfunc='sum', columns='성별').T
```
