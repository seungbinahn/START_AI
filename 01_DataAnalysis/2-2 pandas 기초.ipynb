```python
# group 함수
mygrp = df.groupby('year')
mygrp['lifeExp'].mean()
```

```python
multi_group = df.groupby(['year','continent'])
multi_var = multi_group[['lifeExp','gdpPercap']].mean()
multi_var.head(n=10)
```

```python
#nunique 함수 : 데이터의 개수
multi_group = df.groupby(['year','continent'])
multi_var = multi_group[['lifeExp','gdpPercap']].nunique()
multi_var.head(n=10)
```
