# Regression(회귀) 모델
* 평균 값으로 돌아가려는 경향을 이용한 통계학 기법
* 키 = > 부모의 키가 크다고해서 무조건 자식의 키가 계속해서 커지지는 않음
* 여러 개의 독립변수(Feature, x)와 한 개의 종속변수(Label, y) 간의 상관 관계를 모델링
* 모델링 : 종속변수에 대한 독립변수들의 상관 관계를 회귀 계수(Weight, w)로 나타내고 최적 값을 학습을 통해 찾아내는 것

## 단순 선형 회귀
* 독립변수와 종속변수가 1개씩 존재하는 모델
* Y = W * X + b
* 목표는 W와 b를 찾는 것
* 학습을 통해 차이의 합(RSS)을 최소화 시키는 회귀 계수(W, b)를 찾아 최적의 모델 구현

## RSS(Residual Sum of Squares) 
 = ∑ (실제가격 - 예측가격 ) ^ 2

* H(w,b) = wx + b
* 목표는 w와 b를 찾는 것


## Cost Function
* RSS / 2m 
* loss Function(손실함수), objective function(목적함수), 비용함수 모두 같은 말임

## 경사 하강법(Gradient-descent)
* Cost를 줄이기 위해 각 단계에서의 기울기 계산
* w, b 값을 update
* update 후 오차 계산, 오차가 0이 아니면 다시 위 단계 반복
* 기울기가 음수면 증가하는 방향으로, 양수이면 감소하는 방향으로 진행
